```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pandas as pd


# Sample data and labels (replace with your own data)
qustion_data = pd.read_csv('../data/processed/training_questions.csv')

texts = qustion_data['question']

labels = qustion_data['label']

# Tokenize the text data
tokenizer = Tokenizer(num_words=10000)  # Adjust the vocabulary size as needed
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
word_index = tokenizer.word_index

max_sequence_length = 40  # Adjust as needed
padded_sequences = pad_sequences(sequences, maxlen=max_sequence_length, padding='post', truncating='post')
print(padded_sequences[0])


# Convert labels to one-hot encoding
label_set = set(labels)
label_to_index = {label: idx for idx, label in enumerate(label_set)}
index_to_label = {idx: label for label, idx in label_to_index.items()}
num_classes = len(label_set)


label_indices = [label_to_index[label] for label in labels]
one_hot_labels = tf.keras.utils.to_categorical(label_indices, num_classes=num_classes)



# Split data into training and testing sets
train_size = int(0.8 * len(texts))
train_texts, test_texts = padded_sequences[:train_size], padded_sequences[train_size:]
train_labels, test_labels = one_hot_labels[:train_size], one_hot_labels[train_size:]

# Build the text classification model
model = keras.Sequential([
    Embedding(input_dim=len(word_index) + 1, output_dim=100, input_length=max_sequence_length),
    Bidirectional(LSTM(64)),
    Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


# Train the model
model.fit(train_texts, train_labels, epochs=100, batch_size=32, validation_split=0.2)
```

```python

# # Evaluate the model
loss, accuracy = model.evaluate(test_texts, test_labels)
print(f'Test loss: {loss:.4f}, Test accuracy: {accuracy:.4f}')
# Inference
new_text = "what is the work culture?"  # Replace with your input text
new_text_sequence = tokenizer.texts_to_sequences([new_text])
new_text_padded = pad_sequences(new_text_sequence, maxlen=max_sequence_length, padding='post', truncating='post')
predictions = model.predict(new_text_padded)
predicted_label_index = tf.argmax(predictions, axis=1).numpy()[0]
predicted_label = index_to_label[predicted_label_index]

print(f'Predicted label: {predicted_label}')
```
